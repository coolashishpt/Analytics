```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [23,34,23,26,42,20]
z = [25,43,32,36,41,30]
u = [32,21,38,46,29,34]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, label="max")
plt.plot(x,z, label = "min")
plt.plot(x,u, label = "avg")
plt.legend(loc = "upper right", shadow = True, fontsize = "medium")
plt.grid()
```

```python
company = ["Microsoft", "Facebook", "Amazon", "Google"]
revenue = [90,50,35,103]
profit = [20,30,5,60]
```

```python
arr = np.arange(len(company))
arr
```

```python
# for verticle show
plt.ylabel("Revenue")
plt.title("Stock Analysis in Bar Chart")
plt.xticks(arr, company)
plt.bar(arr-0.2, revenue, width=0.4, label = "Revenue")
plt.bar(arr+0.2, profit, width = 0.4, label = "Profit")
plt.legend()
```

```python
#for horizontal show
plt.ylabel("Revenue")
plt.title("Stock Analysis in Bar Chart")
plt.yticks(arr, company)
plt.barh(arr-0.2, revenue,height = 0.4 , label = "Revenue")
plt.barh(arr+0.2, profit, height = 0.4, label = "Profit")
plt.legend()
```

```python
home_items = ["Food/Veg", "Mobile/Net Bill", "Electricity", "Home Rent", "Traveling", "Others"]
home_items_price = [3000, 2000, 1500, 12000, 1200, 500]
```

```python
plt.pie(home_items_price, labels= home_items, radius = 1.5, shadow=True, autopct="%0.2f%%", explode=[0,0.2,0,0,0.4,0.2])
# for save the chart into png file
plt.savefig("pie.png", bbox_inches = "tight")
```
