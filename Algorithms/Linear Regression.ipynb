```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
from sklearn import datasets
```

```python
data = datasets.load_boston()
print("INFORMATION OF DATA\n")
print(data.keys())
print(37*"--")
print("Features:\n",data.data)
print(37*"--")
print("Label:\n",data.target[10:24])
print(37*"--")
print("Target Name:\n",data.feature_names)
```

```python
print(data.DESCR)
```

```python
x_df = pd.DataFrame(data.data, columns=data.feature_names)
y_df = pd.DataFrame(data.target)
print(x_df.shape)
print(y_df.shape)
x_df[:10]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(np.array(x_df), np.array(y_df), random_state = 0)

print(10*"-"+"TRAIN DATA"+10*"-"+"\n")
print("Shape'X' :",X_train.shape)
print("Shape'Y' :",y_train.shape)
print()
print(10*"-"+"TEST DATA"+10*"-"+"\n")
print("Shape'X' :",X_test.shape)
print("Shape'Y' :",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
print(lr)
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
from sklearn.linear_model import Ridge


R = Ridge(alpha=10)
print(R)
R.fit(X_train, y_train)
print(R.score(X_train, y_train))
print(R.score(X_test, y_test))
```

```python
from sklearn.linear_model import Lasso

L = Lasso()
print(L)
L.fit(X_train, y_train)
print(L.score(X_train, y_train))
print(L.score(X_test, y_test))
```
